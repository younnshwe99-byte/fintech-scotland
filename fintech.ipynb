```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
import time

BASE_URL = "https://www.fintechscotland.com/cluster/fintech-community/"

headers = {
    "User-Agent": "Mozilla/5.0"
}

rows = []

# Main page + pages 1-17
urls = [BASE_URL] + [
    f"{BASE_URL}page/{i}/"
    for i in range(1, 18)
]

for url in urls:
    print(f"Fetching {url}")

    response = requests.get(url, headers=headers, timeout=30)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    for card in soup.select("a.fintech-card"):
        href = card.get("href")
        href = urljoin(BASE_URL, href) if href else None

        h3 = card.find("h3")
        company_name = h3.get_text(strip=True) if h3 else None

        rows.append({
            "company_name": company_name,
            "href": href
        })

    time.sleep(0.5)

# Create DataFrame and remove duplicates
df = (
    pd.DataFrame(rows)
      .drop_duplicates(subset=["href"])
      .reset_index(drop=True)
)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

headers = {
    "User-Agent": "Mozilla/5.0"
}

for i, row in df.iterrows():
    print(f"{i+1}/{len(df)}: {row['company_name']}")

    response = requests.get(row["href"], headers=headers, timeout=30)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    sidebar = soup.select_one("div.fintech-sidebar")

    if sidebar:
        items = sidebar.select("li.overview-info-item")

        for item in items:
            spans = item.find_all("span")

            if len(spans) >= 2:
                label = spans[0].get_text(strip=True)
                value = spans[1].get_text(" ", strip=True)

                # Clean label so it's a nice column name
                label = (
                    label.lower()
                         .replace(" ", "_")
                         .replace("/", "_")
                         .replace(":", "")
                )

                df.loc[i, label] = value

    time.sleep(0.5)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

headers = {
    "User-Agent": "Mozilla/5.0"
}

for i, row in df.iterrows():
    print(f"{i+1}/{len(df)}: {row['company_name']}")

    response = requests.get(row["href"], headers=headers, timeout=30)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    # ---- TAB CONTENT EXTRACTION ----
    tab_container = soup.select_one("div.fintech-tabbed-content")

    if tab_container:
        tabs = tab_container.select("div.tab-content")

        for idx, tab in enumerate(tabs, start=1):
            # extract all text inside tab (including nested divs, p, etc.)
            text = tab.get_text(separator=" ", strip=True)

            col_name = f"tab_{idx}"
            df.loc[i, col_name] = text

    time.sleep(0.5)
```
